```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize = (9,9))
area = [48, 30, 20, 15]
labels = ['Low', 'Medium', 'High', 'Critical']
colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']
plt.pie(area, labels = labels, colors = colors, startangle = 45)
plt.show()
```

```python
plt.figure(figsize = (9,9))
area = [48, 30, 20, 15]
labels = ['Low', 'Medium', 'High', 'Critical']
colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']
plt.pie(area, labels = labels, colors = colors, startangle = 45, shadow = 'true', autopct = '%1.1f%%')
plt.show()
```

```python
plt.figure(figsize = (8,8))
area = [48, 30, 20, 15]
total = np.sum(area)
labels = ['Low', 'Medium', 'High', 'Critical']

def val_per(x):
    return '{:.2f}%\n({:.0f})'.format(x, total*x/100)

colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']
plt.pie(area, labels = labels, colors = colors, startangle = 45, shadow = 'true', autopct = val_per)
plt.show()
```

```python
fig = plt.figure(figsize = (20, 6))
area = [48, 30, 20, 15]
priority = ['Low', 'Medium', 'High', 'Critical']
status = ['Resolved', 'Cancelled', 'Pending', 'Assigned']
company = ['IBM', 'Microsoft', 'Apple', 'Tesla']
colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']

plt.subplot(1, 3, 1)
plt.pie(area, labels = priority, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.subplot(1, 3, 2)
plt.pie(area, labels = status, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.subplot(1, 3, 3)
plt.pie(area, labels = company, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

```python
#Histogram
```

```python
x = np.random.normal(size = 2000)
plt.hist(x, bins = 40, color = 'yellowgreen')
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
x = np.random.rand(2000)
plt.hist(x, bins = 30, color = '#D4AC0D')
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
plt.figure(figsize = (10, 8))
x = np.random.normal(size = 2000)
plt.hist(x, bins = 40, color = 'yellowgreen', edgecolor = '#6A9662')
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
#Binning
```

```python
plt.figure(figsize = (10, 8))
x = np.random.normal(size = 2000)
plt.hist(x, bins = 30, color = 'yellowgreen', edgecolor = '#6A9662')
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()

plt.figure(figsize = (10, 8))
plt.hist(x, bins = 20, color = 'yellowgreen', edgecolor = '#6A9662')
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()

plt.figure(figsize = (10, 8))
plt.hist(x, bins = 10, color = 'yellowgreen', edgecolor = '#6A9662')
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
#multiple histogram
```

```python
plt.figure(figsize = (8, 11))
x = np.random.normal(-4, 1, size = 800)
y = np.random.normal(0, 1.5, size = 800)
z = np.random.normal(3.5, 1, size = 800)
plt.hist(x, bins = 30, color = 'yellowgreen', alpha = 0.6)
plt.hist(y, bins = 30, color = '#FF8F00', alpha = 0.6)
plt.hist(z, bins = 30, color = 'blue', alpha = 0.6)
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
#CDF (Cumulative Distribution Function)
```

```python
plt.figure(figsize = (10,8))
x = np.random.rand(2000)
plt.hist(x, bins = 30, color = '#ffa41b', edgecolor = '#639a67', cumulative = True)
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
#Area Plot
```

```python
x = np.arange(1, 31)
y = np.random.normal(10, 11, size = 30)
y = np.square(y)
plt.figure(figsize = (16, 6))
plt.plot(x, y)
plt.fill_between(x, y)
plt.show()
```

```python
#Stacked Area Plot
```

```python
x = np.arange(1, 6)
y1 = np.array([1,5,9,13,17])
y2 = np.array([2,6,10,14,16])
y3 = np.array([3,7,11,15,19])
y4 = np.array([4,8,12,16,20])

plt.figure(figsize = (8,6))
plt.stackplot(x, y1, y2, y3, y4, labels = ['Y1', 'Y2', 'Y3', 'Y4'])
plt.legend(loc = 'upper left')
plt.show()
```

```python
import seaborn as sns
```

```python
#Line Charts
```

```python
col1 = np.linspace(0, 10, 1000)
col2 = np.sin(col1)
df = pd.DataFrame({'C1' : col1, 'C2' : col2})
df.head(10)
```

```python
plt.style.use('seaborn-darkgrid')
%matplotlib inline
sns.lineplot(x = df.C1, y = df.C2, data = df)
plt.show()
```

```python
plt.figure(figsize = (14,6))
sns.set(rc = {"axes.facecolor" : "#FFFDE7", "axes.grid" : True, 'xtick.labelsize' : 14, 'ytick.labelsize' : 14})
sns.lineplot(x = df.C1, y = df.C2, data = df, linewidth = 2)
sns.lineplot(x = df.C1, y = df.C2 + 1, data = df, linewidth = 2)
plt.show()
```

```python
plt.figure(figsize = (14,6))
sns.set(rc = {"axes.facecolor" : "#283747", "axes.grid" : False, 'xtick.labelsize' : 14, 'ytick.labelsize' : 14})
sns.lineplot(x = df.C1, y = df.C2, data = df, color = "#FF5722", linewidth = 2)
sns.lineplot(x = df.C1, y = df.C2 + 1, data = df, color = "#FFEB3B", linewidth = 2)
plt.show()
```
